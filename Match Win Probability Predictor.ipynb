```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
ball=pd.read_csv('deliveries.csv')
```

```python
match=pd.read_csv('matches.csv')
```

```python
match.head()
```

```python
match['team1'].unique()
```

```python
team=['Sunrisers Hyderabad', 'Mumbai Indians', 'Royal Challengers Bangalore',
       'Kolkata Knight Riders', 'Delhi Daredevils', 'Kings XI Punjab',
       'Chennai Super Kings', 'Rajasthan Royals', 'Deccan Chargers',
       'Delhi Capitals']
```

```python
match=match[match['team1'].isin(team)]
match=match[match['team2'].isin(team)]
```

```python
match.shape
```

```python
match['team1']=match['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match['team1']=match['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match['team2']=match['team2'].str.replace('Delhi Daredevils','Delhi Capitals')
match['team2']=match['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match.head()
```

```python
match=match[['id','city','team1','team2','winner']]
```

```python
ball.head()
```

```python
total_score=ball.groupby(['match_id','inning'])['total_runs'].sum().reset_index()
```

```python
target_score=total_score[total_score['inning'] == 1]
```

```python
match = match.merge(target_score[['match_id','total_runs']],left_on='id',right_on='match_id')
```

```python
match.head()
```

```python
match=match[['match_id','city','winner','total_runs']]
```

```python
ball_df = match.merge(ball,on='match_id')
```

```python
ball_df.head()
```

```python
ball_df=ball_df[ball_df['inning'] == 2]
```

```python
ball_df['current_score'] = ball_df.groupby(['match_id'])['total_runs_y'].cumsum()
```

```python
ball_df.head()
```

```python
ball_df['runs_left'] = ball_df['total_runs_x'] - ball_df['current_score'] 
```

```python
ball_df.head()
```

```python
ball_df['balls_left'] = 126 - (ball_df['over']*6 + ball_df['ball'])
```

```python
ball_df.head()
```

```python
ball_df.tail()
```

```python
# crr = runs/overs
ball_df['crr'] = (ball_df['current_score']*6)/(120 - ball_df['balls_left'])
```

```python
# rrr=runs_left/overs
ball_df['rrr'] = (ball_df['runs_left']*6)/ball_df['balls_left']
```

```python
ball_df.head()
```

```python
ball_df['player_dismissed'] = ball_df['player_dismissed'].fillna("0")
ball_df['player_dismissed'] = ball_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")
ball_df['player_dismissed'] = ball_df['player_dismissed'].astype('int')
wickets = ball_df.groupby('match_id').cumsum()['player_dismissed'].values
ball_df['wickets left'] = 10 - wickets
ball_df.tail()
```

```python
def result(text):
    if text['batting_team'] == text['winner']:
        return 1
    else:
        return 0
```

```python
ball_df['result'] = ball_df.apply(result,axis=1)
```

```python
ball_df.head()
```

```python
final_df = ball_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets left','total_runs_x','crr','rrr','result']]
```

```python
final_df
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df=final_df.rename(columns={'total_runs_x':'target'})
```

```python
final_df.isnull().sum()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df=final_df[final_df['balls_left'] != 0]
```

```python
final_df.head()
```

```python
x=final_df.drop('result',axis=1)
y=final_df['result']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import SGDClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
pipe.predict_proba(x_test)[1]
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',RandomForestClassifier())
])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
pipe.fit(x_train,y_train)
```

```python
accuracy_score(y_test,y_pred)
```

```python
y_pred = pipe.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```
